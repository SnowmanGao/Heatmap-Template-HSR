```python
import json
import os
from datetime import datetime
from time import sleep
from urllib import parse

from bs4 import BeautifulSoup
from requests import Session


def build_url(name: str) -> str:
    fandom = parse.quote("崩坏：星穹铁道 | Honkai: Star Rail (Video Game)")
    name = parse.quote(f'"{name}"')
    return f"https://archiveofourown.org/tags/search?commit=Search+Tags&page=1&tag_search%5Bcanonical%5D=T&tag_search%5Bfandoms%5D={fandom}&tag_search%5Bname%5D={name}&tag_search%5Bsort_column%5D=uses&tag_search%5Bsort_direction%5D=desc&tag_search%5Btype%5D=Relationship&view_adult=true"
```

```python
session = Session()
PATH = "./data/relationships"


def fetch_tags_info(name: str, log=True):
    resp = session.get(build_url(name))
    if log:
        print(resp)
    bs = BeautifulSoup(resp.content)
    total = int(bs.select_one("#main > h3").contents[0].split(" ")[0])
    spans = bs.select("ol.tag span")

    obj = []
    for item in spans:
        chs = list(item.children)
        obj.append({"tag": chs[1].contents[0], "count": int(chs[2][3:-1])})

    if not os.path.exists(PATH):
        os.makedirs(PATH)
    with open(f"{PATH}/{name}.json", "w") as fp:
        json.dump(
            {
                "total": total,
                "number_of_tags": len(spans),
                "date": datetime.now().strftime("%Y-%m-%d %T"),
                "tags": obj,
            },
            fp,
            ensure_ascii=False,
            indent=4,
        )


with open("./SR_NAMES") as fp:
    while name := fp.readline().strip():
        fetch_tags_info(name)
        sleep(1)
```
